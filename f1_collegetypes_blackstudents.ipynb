# Students According to Degree Programs and Ownership of Schools

```stata
use data/hd_efa_2022, clear
```

```stata
drop if control==-3
```

```stata
drop if iclevel==-3
```

```stata
gen collegeyears=.

replace collegeyear=1 if iclevel==1
replace collegeyear=2 if iclevel==2
replace collegeyear=3 if iclevel==3

label define iclevel 1 "4 Year College" 2 "2 Year College" 3 "Less Than 2 Year College"
label values iclevel iclevel
label var collegeyears "Program Duration"
```

```stata
gen ownership=.

replace ownership=1 if control==1
replace ownership=2 if control==2
replace ownership=3 if control==3

label define control 1 "Public" 2 "Private Not for Profit" 3 "Private for Profit"
label values control control
label var ownership "College Ownership Status"
```

```stata
graph bar pctefbkaat, over(control) over(iclevel) asyvars ytitle ("Black Student Population") ///
  blabel(bar, format(%7.0fc)) legend(rows(1) position(1) ring(0)) scheme(538)
```

```stata
graph save Black_Students_Colleges, replace
```

```stata
graph bar pctefwhitt, over(control) over(iclevel) asyvars ytitle ("White Student Population") ///
  blabel(bar, format(%7.0fc)) legend(rows(1) position(1) ring(0)) scheme(538)
```

```stata
graph save White_Students_Colleges, replace
```
